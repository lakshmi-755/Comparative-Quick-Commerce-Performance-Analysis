```python
def categorize_delivery_time(time):
    if time <= 23:
        return 'Fast'
    elif time <= 36:
        return 'Average'
    else:
        return 'Slow'

df['Delivery_Time_Category'] = df['Delivery Time (Minutes)'].apply(categorize_delivery_time)
```

```python
def categorize_customer(value):
    if value <= 283:
        return 'Low-Value Customer'
    elif value <= 770:
        return 'Medium-Value Customer'
    else:
        return 'High-Value Customer'

df['Customer Segment'] = df['Order Value (INR)'].apply(categorize_customer)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)   # remove punctuation
    words = text.split()
    words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words]
    return ' '.join(words)

df['Clean_Feedback'] = df['Customer Feedback'].astype(str).apply(preprocess)
```

```python
positive_words = [
    'fast', 'great', 'excellent', 'good', 'reliable', 'quick',
    'amazing', 'perfect', 'nice'
]

negative_words = [
    'missing', 'delay', 'late', 'damaged', 'bad',
    'worst', 'poor', 'refund', 'issue', 'problem'
]

neutral_words = [
    'could better', 'okay', 'average', 'fine'
]
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

def business_sentiment(text):
    text = text.lower()

    # Rule-based priority
    if any(word in text for word in negative_words):
        return 'Negative'
    
    if any(word in text for word in positive_words):
        return 'Positive'
    
    if any(word in text for word in neutral_words):
        return 'Neutral'

    # Fallback to VADER
    score = sia.polarity_scores(text)['compound']
    if score >= 0.2:
        return 'Positive'
    elif score <= -0.2:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df['Sentiment'] = df['Clean_Feedback'].apply(business_sentiment)
```

```python
sentiment_map = {
    'Negative': -1,
    'Neutral': 0,
    'Positive': 1
}

df['Sentiment_Score'] = df['Sentiment'].map(sentiment_map)
```

```python
df['Delivery Delay_binary'] = df['Delivery Delay'].map({'Yes': 1, 'No': 0})
df['Refund Requested_binary'] = df['Refund Requested'].map({'Yes': 1, 'No': 0})
```

```python
features_churn = df[
    ['Delivery Delay_binary', 'Refund Requested_binary', 'Service Rating', 'Sentiment_Score']
].astype(int)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_churn)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


# Apply K-Means with 2 clusters
kmeans_churn = KMeans(
    n_clusters=2,
    random_state=42,
    n_init=10
)

df['Churn_Risk_Cluster'] = kmeans_churn.fit_predict(features_scaled)
```

```python
cluster_summary = df.groupby('Churn_Risk_Cluster')[[
    'Delivery Delay_binary',
    'Refund Requested_binary',
    'Service Rating',
    'Sentiment_Score'
]].mean()

print(cluster_summary)
```

```python
cluster_map = {
    0: 'High Risk',
    1: 'Low Risk'
}

df['Churn_Risk_Segment'] = df['Churn_Risk_Cluster'].map(cluster_map)
```

```python
features_delay = df[
    ['Delivery Time (Minutes)', 'Service Rating', 'Sentiment_Score']
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_delay_scaled = scaler.fit_transform(features_delay)
```

```python
from sklearn.cluster import KMeans

kmeans_delay = KMeans(
    n_clusters=2,
    random_state=42,
    n_init=10
)

df.loc[features_delay.index, 'Delay_Sensitivity_Cluster'] = (
    kmeans_delay.fit_predict(features_delay_scaled)
)
```

```python
delay_cluster_summary = df.groupby('Delay_Sensitivity_Cluster')[[
    'Delivery Time (Minutes)',
    'Service Rating',
    'Sentiment_Score'
]].mean()

print(delay_cluster_summary)
```

```python
cluster_map_delay = {
    0: 'Delay-Sensitive',
    1: 'Delay-Tolerant'
}

df['Delay_Sensitivity_Segment'] = (
    df['Delay_Sensitivity_Cluster'].map(cluster_map_delay)
)
```
