```python
#Example 1: Platform vs Delivery Time
groups = [
    df[df['Platform'] == platform]['Delivery Time (Minutes)']
    for platform in df['Platform'].unique()
]

f_stat, p = f_oneway(*groups)
print("\nANOVA: Platform vs Delivery Time")
print(f"F-statistic = {f_stat:.3f}, p-value = {p:.4f}")
```

```python
#Example 2: Delivery Delay vs Service Rating
groups = [
    df[df['Delivery Delay'] == delay]['Service Rating']
    for delay in df['Delivery Delay'].unique()
]

f_stat, p = f_oneway(*groups)
print("\nANOVA: Delivery Delay vs Service Rating")
print(f"F-statistic = {f_stat:.3f}, p-value = {p:.4f}")
```

```python
#Example 3: Product Category vs Order Value
groups = [
    df[df['Product Category'] == cat]['Order Value (INR)']
    for cat in df['Product Category'].unique()
]

f_stat, p = f_oneway(*groups)
print("\nANOVA: Product Category vs Order Value")
print(f"F-statistic = {f_stat:.3f}, p-value = {p:.4f}")
```
